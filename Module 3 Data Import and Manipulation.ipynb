# Module 3: Data Import and Manipulation

## 1. Loading Data into Jupyter Notebook:

In this section, we'll learn how to load data into Jupyter Notebook. We'll use Pandas to read a CSV file as an example.

```python
# 1. Loading Data into Jupyter Notebook

# To load data into Jupyter Notebook, we can use Pandas, a powerful data manipulation library.

# Import Pandas
import pandas as pd

# Let's load a sample CSV file into a Pandas DataFrame.
# You can replace 'sample_data.csv' with the path to your own data file.
df = pd.read_csv('sample_data.csv')

# Display the first few rows of the DataFrame to explore the data
df.head()
```

## 2. Exploring and Cleaning Data with Pandas:

In this section, we'll explore and clean the loaded data using Pandas.

```python
# 2. Exploring and Cleaning Data with Pandas

# Now that we've loaded data, let's explore and clean it.

# Check the shape of the DataFrame (number of rows and columns)
print("Shape of the DataFrame:", df.shape)

# Check the data types of columns
print("Data Types of Columns:")
print(df.dtypes)

# Basic statistics of numeric columns
print("Summary Statistics:")
print(df.describe())

# Check for missing values
print("Missing Values:")
print(df.isnull().sum())

# Drop rows with missing values (use with caution)
df_cleaned = df.dropna()

# Explore the cleaned DataFrame
print("Cleaned DataFrame:")
print(df_cleaned.head())

# Now, let's move on to data visualization.
```

## 3. Data Visualization using Matplotlib and Seaborn:

In this section, we'll visualize the data using Matplotlib and Seaborn for data analysis.

```python
# 3. Data Visualization using Matplotlib and Seaborn

# Data visualization is a crucial part of data analysis. We'll use Matplotlib and Seaborn for this purpose.

# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Let's create a simple scatter plot to visualize data
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cleaned, x='Feature1', y='Feature2')
plt.title('Scatter Plot of Feature1 vs. Feature2')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()

# You can customize the plot further as needed.

# These are the essential steps for loading, exploring, cleaning, and visualizing data in Jupyter Notebook.

```
